Write a function which implements PCA via the eigendecomposition. <b>(10 points)</b>

You will be given as input:
- A $(N,d)$ numpy array of data (with each row as a feature vector)

Your function should return a tuple consisting of the PCA transformation matrix (which is $(d,d)$), and a vector of length $d$ consisting of the amount of variance explained in the data by each PCA feature. Note that the PCA features are ordered in decreasing amount of variance explained.

Hints:
- The function <a href="http://docs.scipy.org/doc/numpy-1.10.0/reference/generated/numpy.linalg.eigh.html">numpy.linalg.eigh</a> will be useful. Note that it returns its eigenvalues in *ascending* order. 
- You can calculate the covariance matrix of the data by multiplying the data matrix with its transpose in the appropriate order, and scaling it. 
- Do not use numpy.cov -- we are assuming the data has zero mean beforehand, so the number of degrees of freedom is different (since the covariance estimate knows the mean in our case). 

```python
def pcaeig(data):
    cov = np.dot(np.transpose(data),data)
    cov = np.divide(cov,data.shape[0])
    eigh = np.linalg.eigh(cov)
    eig = eigh[0][::-1]
    eigV = np.fliplr(eigh[1])
    return np.transpose(eigV),eig
```

Now, run PCA on your data via `pcaeig`. Store your PCA transformation in a variable called `W`, and the amount of variance explained by each PCA feature in a variable called `s`. Print out the principal components along with the corresponding amount of variance explained. <b>(5 points)</b>

```python
# Put your code here
result = pcaeig(data)
W = result[0]
s = result[1]
print "Principal Components:"
print W
print "\nVariances:"
print s
```

Write a function which implements dimension reduction via PCA. It takes in three inputs:
- A $(N,d)$ numpy array, `data`, with each row as a feature vector
- A $(d,d)$ numpy array, `W`, the PCA transformation matrix (e.g. generated from `pcaeig` or `pcasvd`)
- A number `k`, which is the number of PCA features to retain

It should return a $(N,k)$ numpy array, where the $i$-th row contains the PCA features corresponding to the $i$-th input feature vector ($i$-th row of `data`). <b>(5 points)</b>

```python
def pcadimreduce(data,W,k):
    c1 = np.dot(W[0:k],data.T)
    return np.einsum('ji,ki->kj',W[0:k],data)
```

Write a function which reconstructs the original features from the PCA features. It takes in three inputs:
- A $(N,k)$ numpy array, `pcadata`, with each row as a PCA feature vector (e.g. generated from `pcadimreduce`)
- A $(d,d)$ numpy array, `W`, the PCA transformation matrix (e.g. generated from `pcaeig` or `pcasvd`)
- A number `k`, which is the number of PCA features

It should return a $(N,d)$ numpy array, where the $i$-th row corresponds to the reconstruction in the original feature space of the $i$-th row of PCA features contained in `pcadata`. <b>(5 points)</b>

```python
def pcareconstruct(pcadata,W,k):
    c1 = np.dot(W[0:k].T,pcadata.T)
    return np.einsum('ji,ki->kj',W[0:k].T,pcadata)
```

Finally, you will implement PCA via the SVD. <b>(5 points)</b>

Your function should return a tuple consisting of the PCA transformation matrix (which is $(d,d)$), and a vector of length $d$ consisting of the amount of variance explained in the data by each PCA feature. Note that the PCA features are ordered in decreasing amount of variance explained.

Hints:
- The function <a href="http://docs.scipy.org/doc/numpy-1.10.0/reference/generated/numpy.linalg.svd.html">numpy.linalg.svd</a> will be useful. Use the full SVD (default).
- Be careful with how the SVD is returned in `numpy.linalg.svd` (`V` in numpy is the transpose of what is in the notes). 

```python
def pcasvd(data):
    svd = np.linalg.svd(data)
    return svd[2],np.divide(np.power(svd[1],2),data.shape[0])
```

```python
#Put your code here
def classifierError(truelabels,estimatedlabels):
    return float(np.sum(truelabels!=estimatedlabels))/truelabels.size

feats = list(range(1,257))
mailPCA = PCA(256)
mailPCA.fit(traindata)
err = np.zeros(len(feats))
TRcomps = np.copy(mailPCA.components_)

for i in xrange(0,len(feats)):
    mailPCA.components_[feats[i]:] = 0   
    trainTransform = mailPCA.transform(traindata)
    invTransform = mailPCA.inverse_transform(trainTransform)
    classifier = neighbors.KNeighborsClassifier(1)
    train = classifier.fit(invTransform,trainlabels)
    vals = train.predict(valdata)
    err[i] = classifierError(vallabels,vals)
    mailPCA.components_ = np.copy(TRcomps)
y = err   
x = feats
min_error =  np.amin(err)
min_error_arg = np.argmin(err)
print ('Min Error: %f' % min_error)
print ('K = %d' % feats[min_error_arg]) 
plot(x,y)
```
